En esta sección filtramos las columnas de las tablas para que sean solo numericas

```python
datos_ml = datos_ml.fillna(0)
datos_ml_numeric = datos_ml.select_dtypes(include=['number'])
columnas_posibles = [col for col in datos_ml_numeric.columns if col != 'Precio_Millon']
columnas_posibles
```

importamos la biblioteca itertools para generar todas las combinaciones posibles de las columnas que podrían incluirse en el modelo.

```python
import itertools

mejor_r2 = 0
mejores_columnas = []

# Bucle para ir agregando columnas
for i in range(1, len(columnas_posibles) + 1):
  for columnas in itertools.combinations(columnas_posibles, i):
        X = datos_ml[list(columnas)]
        y = datos_ml["Precio_Millon"]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=99)

        modelo = LinearRegression()
        modelo.fit(X_train, y_train)

        y_predict_test = modelo.predict(X_test)
        r2_test = r2_score(y_test, y_predict_test)

        if r2_test > mejor_r2:
            mejor_r2 = r2_test
            mejores_columnas = columnas

        # Detener si el r2 es mayor que un cierto umbral (por ejemplo, 0.7)
        if mejor_r2 > 0.7:
            break

    # Detener si el r2 es mayor que un cierto umbral (por ejemplo, 0.7)
  if mejor_r2 > 0.7:
      break

print(f"Mejor R2: {mejor_r2}")
print(f"Mejores columnas: {mejores_columnas}")
```
