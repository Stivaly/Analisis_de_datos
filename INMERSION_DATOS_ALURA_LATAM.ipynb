Acá realizamos la limpieza de la columna "Valor" la cual pudimos ver 2 celdas atrás que corresponde a una cadena, lo que imposibilita poder realizar cálculos.

Para ello realizaremos la división en dos de la cadena con el método .split, luego asignaremos una parte de la cadena a una columna diferente, en este caso Moneda contendrá el símbolo peso y Precio tendrá los números.

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

A continuación la columna precio tiene más caracteres que no nos permiten sacar cálculos como los puntos, por lo que utilizamos el método replace para reemplazar el punto por nada.

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio','Barrio']]
```

En esta sección filtramos las columnas de las tablas para que sean solo numericas y convertimos los campos núlos a cero.

```python
datos_ml = datos_ml.fillna(0)
datos_ml_numeric = datos_ml.select_dtypes(include=['number'])
columnas_posibles = [col for col in datos_ml_numeric.columns if col != 'Precio_Millon']
```

A continuación realizamos la construcción del modelo, en esta parte optimicé el codigo para poder llegar al modelo correcto más rápidamente.

Durante la inmersión de datos en Alura Latam realizamos estos pasos de manera manual, por lo que Alura Latam se encargó del procesamiento de los datos y nosotros solo debíamos ir probando variables e ir experimentando hasta encontrar un modelo apropiado.

En un inicio los resultados de r2 nos dieron entre 0,01 y 0,07 por lo que sabíamos que los modelos no eran apropiados y solo estabamos en proceso de aprendizaje.

Decidí aceerar este proceso agregando un ciclo For que me permitiese ir probando modelos y analizando hasta encontra la mejor combinación para la predicción, de acá hacia adelante veremos los resultados.

```python
mejor_r2 = 0
mejores_columnas = []

# Bucle para ir agregando columnas
for columna in columnas_posibles:
    # Agregar la nueva columna a las columnas existentes
    columnas_a_probar = mejores_columnas + [columna]

    # Dividir el conjunto de datos
    X = datos_ml[columnas_a_probar]
    y = datos_ml["Precio_Millon"]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=99)

    # Entrenar el modelo
    modelo = LinearRegression()
    modelo.fit(X_train, y_train)

    # Realizar predicciones y calcular R^2
    y_pred = modelo.predict(X_test)
    r2_actual = r2_score(y_test, y_pred)

    # Comparar el rendimiento con el mejor R^2 encontrado hasta ahora
    if r2_actual > mejor_r2:
        mejor_r2 = r2_actual
        mejores_columnas = columnas_a_probar
        print(f"Nuevo mejor R^2 = {mejor_r2} con columnas {mejores_columnas}")
    else:
        print(f"Sin mejora con la columna {columna}. Mejor R^2 actual = {mejor_r2}")

# Mostrar las mejores columnas encontradas
print("Mejores columnas encontradas: ", mejores_columnas)
```

En esta parte entrenamos el modelo con las columnas que pudimos tomar de nuestro algortimo anterior y podemos ver que la precisión de prueba ha subido aun 61%, sabemos que aún falta pero el número es mucho más óptimo ahora

```python
X = datos_ml[mejores_columnas]
Y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```
